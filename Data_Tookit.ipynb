1- What is the difference between multithreading and multiprocessing?
- Multithreading :- Multithreading involves the concurrent execution of multiple threads within a single process. Each thread
represents an independent flow of execution, capable of performing tasks simultaneously with other threads.
Threads within the same process share the same data space and system resources, allowing them to
communicate and coordinate effectively.<br>
  - Key Features
    - Runs inside one process
    - Threads share memory
    - Lower overhead (lighter and faster to create)
    - Good for I/O-bound tasks (waiting operations)

- Multiprocessing :- Multiprocessing is a programming and execution model that involves the concurrent execution of multiple
processes. A process is an independent program that runs in its own memory space and has its own resources.
In multiprocessing, multiple processes run concurrently, each with its own set of instructions and data. These
processes can communicate with each other through inter-process communication (IPC) mechanisms.<br>
  - key Features
    - Runs separate processes
    - Each process has its own memory
    - Higher overhead (heavier than threads)
    - True parallelism on multi-core CPUs

2- What are the challenges associated with memory management in Python?
- Python provides automatic memory management using **reference counting** and a **garbage collector**, which reduces the need for manual memory allocation and deallocation. However, several challenges still exist:
  1. **Reference Cycles** :-
   When objects reference each other, their reference count never becomes zero. Such circular references are not freed immediately and depend on the garbage collector, which may delay cleanup and increase memory usage.

  2. **Memory Leaks**
   Memory leaks can still occur due to global variables, long-lived data structures, improper caching, or objects not being released by external libraries (C extensions).

  3. **Memory Fragmentation**
   Freed memory is not always returned to the operating system. This leads to fragmentation and higher apparent memory consumption, especially in long-running programs.

  4. **Object Overhead**
   Python objects have additional metadata (type info, reference count), so they consume more memory compared to low-level languages. Handling large datasets can therefore be memory intensive.

  5. **Garbage Collection Overhead**
   Garbage collection takes time and can cause small execution pauses, which may affect performance in real-time or high-performance applications.

  6. **Hidden Copies of Data**
   Some operations (like slicing, concatenation, and certain library operations) create copies of objects, leading to unexpected memory usage.

  7. **External Library Issues**
   Libraries written in C/C++ may not follow Python’s garbage collection properly and can cause memory leaks if not managed carefully.

6-  What are the differences between NumPy arrays and Python lists?
- Differences Between NumPy Arrays and Python Lists:-

  1. **Library Requirement**

     * Python lists are built-in data structures.
     * NumPy arrays require the **NumPy** library.

  2. **Data Type**

     * Lists can store elements of **different data types** (int, float, string, etc.).
     * NumPy arrays usually store elements of the **same data type**.

  3. **Performance**

     * Lists are slower for numerical computations.
     * NumPy arrays are **faster and more efficient** for mathematical operations.

  4. **Memory Usage**

     * Lists use more memory because they store references to objects.
     * NumPy arrays use **less memory** due to fixed data type storage.

  5. **Mathematical Operations**

     * Lists do not support element-wise operations directly (need loops).
     * NumPy arrays support **vectorized operations** (e.g., array + array).

  6. **Multidimensional Support**

     * Lists can be nested but are not optimized for multi-dimensional data.
     * NumPy arrays have built-in **multi-dimensional array** support.

  7. **Use Case**

     * Lists are best for general-purpose storage.
     * NumPy arrays are best for **scientific and numerical computing**.


7- Explain the difference between apply() and map() in Pandas.
- Difference Between apply() and map() in Pandas :-

  1. map()

     * Used only on a **Series (single column)**.
     * Performs **element-wise operations**.
     * Accepts a function, dictionary, or another Series.
     * Suitable for simple transformations.

        Example:

        df['A'] = df['A'].map(lambda x: x * 2)

  2. apply()

     * Used on both **Series and DataFrame**.
     * Can perform **row-wise or column-wise operations** (using axis in DataFrame).
     * More flexible and suitable for complex operations.

       Example:

       df['Total'] = df.apply(lambda row: row['A'] + row['B'], axis=1)

```python
# 10-  Calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

# Calculate correlation matrix
corr_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

# Add title
plt.title("Correlation Matrix Heatmap")

plt.show()
```
